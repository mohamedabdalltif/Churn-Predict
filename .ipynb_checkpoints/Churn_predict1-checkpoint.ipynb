```python
df['seniorcitizen'] = df['seniorcitizen'].astype('object')
categorical = list(df.select_dtypes(include=['object']).columns)
numerical = list(df.select_dtypes(include=['number']).columns)
```

```python
# risk ratio
"""
if risk == 1 there is no risk and this means this feature is not important
if risk > 1 there is  a risk that this people more churn than others and this means this feature is  important
if risk < 1 there is  a risk that this people less churn than others and this means this feature is  important

"""
from IPython.display import display

global_mean = df.churn.mean()
round(global_mean, 2)
for feature in categorical:                                           
    df_group = df.groupby(by=feature).churn.agg(['mean']) 
    df_group['diff'] = df_group['mean'] - global_mean
    df_group['risk'] = df_group['mean'] / global_mean
    display(df_group)
```

```python
for feature in categorical:
    df_group=df.groupby(by=feature).churn.agg(['mean']).reset_index()
    graph=sns.barplot(x= feature, y = 'mean', data=df_group, palette='Greens')
    graph.axhline(global_mean, linewidth=2, color='b')
    plt.text(0, global_mean - 0.02, "global_mean", color='black', weight='semibold')
    plt.show()
```

```python
from sklearn.metrics import mutual_info_score

def calculate_mi(series):
    return mutual_info_score(series, df.churn)

df_mi = df[categorical].apply(calculate_mi)
df_mi = df_mi.sort_values(ascending=False).to_frame(name='MI')
less_important_feature1=list((df_mi[df_mi["MI"]<0.019194]).index)
less_important_feature2=list((df_mi[df_mi["MI"]<0.014467]).index)
less_important_feature3=list((df_mi[df_mi["MI"]<0.011454]).index)
less_important_feature4=list((df_mi[df_mi["MI"]<0.010577]).index)
display(df_mi)
###
"""
if value of mi is high this means that this feature is importance 
else not 
"""
```

```python
df[numerical].corrwith(df.churn)
```

```python
for feature in numerical:
    anova_args = tuple(df.groupby('churn')[feature].apply(list).reset_index()[feature])
    f_statistic, p_value = stats.f_oneway(*anova_args)
    if(p_value<0.05):
        print(feature,"p_value is less than alph that = 0.05 and this means that we reject null-hypothesis and it means features do have a dependence between them\n")
    else:
        print(feature,"p_value is more than alph that = 0.05 and this means that we accept null-hypothesis and it means features do not have a dependence between them\n")

    
```

```python
lb = LabelEncoder()
scaler = StandardScaler()
X=df[numerical+categorical]
lst = X[categorical]
for i in lst:
    X[i] = lb.fit_transform(X[i].astype(str))

# df_feature.head()
X[numerical]=scaler.fit_transform(X[numerical])
y=df['churn']
print(X)
```

```python
X.drop(columns=less_important_feature1,axis=1,inplace=True)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=41)
```

```python
from sklearn.linear_model import LogisticRegression
model_1 = LogisticRegression(solver='liblinear', random_state=41)
model_1.fit(X_train, y_train)
print('LogisticRegression Training Accuracy: ', round(model_1.score(X_train, y_train), 2))

print('LogisticRegression Testing Accuracy: ', round(model_1.score(X_test, y_test), 2))
y_pred = model_1.predict(X_test)
```
